```python
df= spark.table('adidas_us_sales_datasets')
display(df)
```

```python
from pyspark.sql.functions import expr

# Example: Use try_cast to tolerate malformed input and return NULL instead
df_changed = df.withColumn('Total_Sales', expr("try_cast(regexp_replace(Total_Sales, '[^0-9]', '') as int)")) \
               .withColumn('Operating_Profit', expr("try_cast(regexp_replace(Operating_Profit, '[^0-9]', '') as int)"))

display(df_changed)
```

```python
df_changed.createOrReplaceTempView("sample")
```

```python
%sql
select sum(Total_Sales) as total_sales,sum(Operating_Profit) as total_profit,avg(price_per_unit) as avg_price,sum(units_sold)as total_units_sold from sample
```

```python
%sql
select sum(Total_Sales) as total_sales, date_format(Invoice_Date,"MMMM") as invoice
from sample
group by Invoice_Date
```

```python
%sql
select sum(Total_Sales) as total_sales,state
from sample
group by state
order by 1 desc 
limit 5
```

```python
%sql
select sum(Total_Sales) as total_sales,region
from sample
group by region
```

```python
%sql
select sum(Total_Sales) as total_sales,product
from sample
group by product
order by 1 desc
limit 5
```

```python
%sql
select sum(Total_Sales) as total_sales,retailer
from sample
group by 2
order by 1 desc
limit 5
```

```python
%sql
select sum(units_sold) as units_sold,sales_method
from sample
group by 2
order by 1 desc
limit 5
```

```python
%sql
select sum(operating_profit) as total_profit ,city
from sample
group by 2
order by 1 desc
limit 5
```
